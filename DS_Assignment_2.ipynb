Write a Pandas program to compare the elements of the two Pandas Series

```python
x = pd.Series([11, 13, 15, 17, 19])
y = pd.Series([11, 13, 15, 18, 19])
comparison = x.eq(y)
print(f"Are x and y equal? \n{comparison}")
```
